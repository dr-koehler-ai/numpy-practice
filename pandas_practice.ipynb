```python
import pandas as pd

patients_df = pd.DataFrame({
    "Patient_ID": [1, 2, 3, 4, 5],
    "HR": [80, 110, 130, 90, 150],
    "RR": [120, 85, 70, 95, 60],
    "Lactate": [1.0, 2.5, 3.0, 0.8, 4.2]
})

patients_df["Shock_Index"] = patients_df["HR"] / patients_df["RR"]

#Gib den DataFrame aus.
print(patients_df)

#Welcher Patient hat den höchsten Shock Index?
print(patients_df.Shock_Index.max())

#Berechne den durchschnittlichen Shock Index.
print(patients_df.Shock_Index.mean())

#Zeige alle Patienten mit HR > 100.
print(patients_df[patients_df.HR > 100])

#Zeige alle Patienten mit RR < 90.
print(patients_df[patients_df.RR < 90])

#Zeige alle Patienten mit Lactate > 2.
print(patients_df[patients_df.Lactate > 2])

#Zeige alle Patienten mit HR > 100 und Lactate > 2
print(patients_df[(patients_df.Lactate > 2) & (patients_df.HR > 100)])
```

```python
import numpy as np

np.random.seed(42)

patients = pd.DataFrame({
    "HR": np.random.randint(60,160,100),
    "RR": np.random.randint(50,140,100),
    "Lactate": np.random.uniform(0,8,100)
})

#Erstelle eine neue Spalte:
patients["Risk_Score"] = 0

high_rr = patients["HR"] > 100
low_rr = patients["RR"] < 90
high_lactate = patients["Lactate"] > 2

score = high_rr.astype(int) + low_rr.astype(int) + high_lactate.astype(int)

patients["Risk_Score"] = score

def risk_class(score):
    if score <= 1:
        return "low"
    elif score == 2:
        return "medium"
    else:
        return "high"

patients["Risk_Class"] = [risk_class(s) for s in patients["Risk_Score"]]

#Sortiere nach Score.
patients.sort_values("Risk_Score", ascending=False)

#Gib alle Patienten mit Score ≥2 aus.
print(patients[patients.Risk_Score > 1])

#Wie hoch ist deren mittleres Lactat?
high_risc_score = patients[patients.Risk_Score > 1]
print(high_risc_score.Lactate.mean())
```
